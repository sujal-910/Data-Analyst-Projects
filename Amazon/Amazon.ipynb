* Pandas library is used to read, view, and manipulate datasets in tabular form.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\sujal class\Data Analysis project\Amazon\amazon.csv")
```

```python
df.head()
```

Column Description :-
* product_id :- This column contains a unique identification number for each product.
* product_name :- This column shows the name of the product listed on Amazon.
* category :- This column represents the category to which the product belongs.
* discounted_price :- This column contains the price of the product after discount.
* actual_price :- This column shows the original price of the product before discount.
* discount_percentage :- This column indicates the percentage discount offered on the product.
* rating :- This column represents the average customer rating for the product.
* rating_count :- This column shows how many users have rated the product.
* review_title :- This column contains the title of the customer review.
* review_content :-This column includes the detailed review written by the customer.

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
df.head()
```

```python
df['rating_type'] = df['rating'].apply(lambda x: 'Good' if x >= 4 else 'Average')
```

```python
df[['rating', 'rating_type']].head()
```

```python
category_dummies = pd.get_dummies(df['category'], drop_first=True)
```

```python
category_dummies.head()
```

```python
df = pd.concat([df, category_dummies], axis=1)
```

```python
df.head()
```

```python
plt.hist(df['discounted_price'])
plt.xlabel("Discounted Price")
plt.ylabel("Number of Products")
plt.title("Discounted Price Distribution")
plt.show()
```

```python
df['rating_type'].value_counts().plot(kind='bar')
plt.xlabel("Rating Type")
plt.ylabel("Number of Products")
plt.title("Distribution of Product Rating Types")
plt.show()
```

-This graph represents how products are distributed based on their ratings.
-It shows whether more products are rated Good or Average by customers.
-From this graph, we can understand the overall customer satisfaction level of the products.
-If the Good category has more products, it means most products are well received by users.
