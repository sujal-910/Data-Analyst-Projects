# Importing Libraries

* Pandas library is used to read ,view and  manipulate datasets in tabular form.
* NumPy is used for mathematical operations and numerical calculations.
* Matplotlib is used for data visualization to analyze distributions, trends, and outliers during exploratory data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['AvgMonthlySpend'] = df['TotalCharges'] / (df['tenure'] + 1)
```

```python
df.head()
```

```python
df['LoyalCustomer'] = np.where(df['tenure'] > 24, 1, 0)
```

```python
df.head()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
plt.hist(df['MonthlyCharges'], bins=30)
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")
plt.title("Distribution of Monthly Charges")
plt.show()
```

```python
df['Churn'].value_counts().plot(kind='bar')
plt.xlabel("Churn Status")
plt.ylabel("Number of Customers")
plt.title("Customer Churn Count")
plt.show()
```
