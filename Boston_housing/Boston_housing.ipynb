Column Descriptions:-
* crim – Crime rate per capita in the area
* zn – Percentage of residential land zoned for large houses
* indus – Proportion of non-retail business areas
* chas – Indicates whether the house is near the Charles River (1=yes, 0=no)
* nox – Level of air pollution (nitric oxide concentration)
* rm – Average number of rooms per house
* age – Percentage of houses built before 1940
* dis – Distance to major employment centers
* rad – Accessibility to highways
* tax – Property tax rate
* ptratio – Pupil-teacher ratio in schools
* lstat – Percentage of lower-status population
* medv – Median value of owner-occupied houses (target variable)

```python
plt.hist(df['medv'],bins=20)
plt.xlabel("Median House Value (MEDV)")
plt.ylabel("Frequency")
plt.title("Distribution of House Prices")
plt.show()
```

```python
plt.scatter(df['rm'], df['medv'])
plt.xlabel("Average Number of Rooms (RM)")
plt.ylabel("Median House Price (MEDV)")
plt.title("Relationship Between Number of Rooms and House Price")
plt.show()
```

```python
df['price_per_room']=df['medv']/df['rm']
```
