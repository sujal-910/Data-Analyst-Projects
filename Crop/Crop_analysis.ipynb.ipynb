## Feature Engineering

```python
df['N_P_ratio'] = df['N'] / (df['P'] + 1)
```

```python
df.head()
```

```python
df['P_K_ratio'] = df['P'] / (df['K'] + 1)
```

```python
df['N_K_ratio'] = df['N'] / (df['K'] + 1)
```

```python
df.head()
```

```python
df['total_nutrients'] = df['N'] + df['P'] + df['K']
```

```python
df.head()
```

```python
df['temp_humidity'] = df['temperature'] * df['humidity']
```

## Data Visualization

```python
plt.hist(df['total_nutrients'], bins=20)
plt.title('Distribution of Total Nutrients')
plt.xlabel('Total Nutrients')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['rainfall'], bins=20)
plt.title('Distribution of Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.show()
```
